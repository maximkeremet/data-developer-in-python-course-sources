2) –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏: –ó–∞—Ç–µ–º –æ–Ω –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç –≤—ã—Ä–∞–∂–µ–Ω–∏–µ a + b, –≤—ã—á–∏—Å–ª—è–µ—Ç —Å—É–º–º—É –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç (30) –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π —Å—É–º–º–∞.
3) –í—ã–≤–æ–¥: –ù–∞–∫–æ–Ω–µ—Ü, –æ–Ω –∏—Å–ø–æ–ª–Ω—è–µ—Ç –∫–æ–º–∞–Ω–¥—É print(—Å—É–º–º–∞), –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π —Å—É–º–º–∞ –Ω–∞ —ç–∫—Ä–∞–Ω.
